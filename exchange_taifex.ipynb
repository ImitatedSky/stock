```python
Date = [data[i]['Date'] for i in range(len(data))] # x axis
USD = [float(data[i]['USD/NTD']) for i in range(len(data))]
RMB = [float(data[i]['RMB/NTD']) for i in range(len(data))]
EUR = [float(data[i]['EUR/USD'])*float(data[i]['USD/NTD']) for i in range(len(data))]
JPY = [float(data[i]['USD/NTD'])/float(data[i]['USD/JPY']) for i in range(len(data))]
GBP = [float(data[i]['GBP/USD'])*float(data[i]['USD/NTD']) for i in range(len(data))]
AUD = [float(data[i]['AUD/USD'])*float(data[i]['USD/NTD']) for i in range(len(data))]
HKD = [float(data[i]['USD/NTD'])/float(data[i]['USD/HKD']) for i in range(len(data))]
RMB = [float(data[i]['USD/NTD'])/float(data[i]['USD/RMB']) for i in range(len(data))]
ZAR = [float(data[i]['USD/NTD'])/float(data[i]['USD/ZAR']) for i in range(len(data))]
NZD = [float(data[i]['NZD/USD'])*float(data[i]['USD/NTD']) for i in range(len(data))]

plt.plot(Date, USD, label = "USD")
plt.plot(Date, RMB, label = "RMB")
plt.plot(Date, EUR, label = "EUR")
plt.plot(Date, JPY, label = "JPY")
plt.plot(Date, GBP, label = "GBP")
plt.plot(Date, AUD, label = "AUD")
plt.plot(Date, HKD, label = "HKD")
plt.plot(Date, RMB, label = "RMB")
plt.plot(Date, ZAR, label = "ZAR")
plt.plot(Date, NZD, label = "NZD")

plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Exchange Rate to NTD')
plt.legend()

plt.show()
```
